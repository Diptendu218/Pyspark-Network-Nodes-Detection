```python
# Define the data
x_out_degree = [0, 1, 2, 5, 8, 26]
y_out_degree = [14, 5, 3, 1, 1, 1]

# Calculate the cumulative sum of y_out_degree
y_cumulative = [sum(y_out_degree[:i+1]) for i in range(len(y_out_degree))]
x_cumulative = [sum(x_out_degree[:i+1]) for i in range(len(x_out_degree))]
# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(y_cumulative,x_cumulative,  align='center', alpha=0.5)

# Add axis labels and title
plt.xlabel('Number of nodes')
plt.ylabel('Number of edges')
plt.title('Cumulative distribution of node degrees')

# Show the plot
plt.show()
```
